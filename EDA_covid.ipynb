```python
from sklearn.preprocessing import MinMaxScaler
def normalized_data(data):
  scaler = MinMaxScaler(feature_range=(0, 1))
  ndata = scaler.fit_transform(np.array(data['new_cases']).reshape(-1,1))
  n_data = ndata.reshape(-1)
  return n_data
    
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
def vac_graph(data,country):
  #secondary y axis is taken for multiple y axis
  fig = make_subplots(specs=[[{'secondary_y':True}]])
  #calculating impordent ratio
  booster_ratio = data['total_boosters']/data['population']
  vaccine_ratio = data['people_fully_vaccinated']/data['population']
  new_case_ratio = normalized_data(data)
  new_df = pd.DataFrame({'date':data['date'],'booster_ratio':booster_ratio,'vaccine_ratio':vaccine_ratio,'new_case_ratio':new_case_ratio})
  fig.add_trace(
      go.Scatter(x=new_df['date'],y=new_df['new_case_ratio'],name='New_case')
  )
  fig.add_trace(
      go.Scatter(x=new_df['date'],y=new_df['vaccine_ratio'],name='Vaccine Ratio')
  )
  fig.add_trace(
      go.Scatter(x=new_df['date'],y=new_df['booster_ratio'],name='Booster Ratio')
  )
  fig.update_layout(title=f'Boost shot & Fully Vaccienate Efficient ? Covid-19 in {country}',
                     xaxis_title='Date',
                     yaxis_title='Number of Cases ')
  fig.show()
```

```python


from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi = make_mi_scores(X_usa,y_usa, discrete_features= True)
mi[::3]
```

```python
mi
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")



```

```python

plt.figure(dpi=100, figsize=(4, 8))
plot_mi_scores(mi)
```
